```python
import os
import cv2

def crop_and_save_melspectrograms(input_dir, output_dir, crop_coordinates):
    """
    Crop mel spectrograms from images and save them to a new directory.

    Parameters:
    - input_dir: Directory containing the original images.
    - output_dir: Directory where cropped images will be saved.
    - crop_coordinates: Tuple (x, y, width, height) defining the crop area.
    """
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Get a list of all image files in the input directory
    image_files = [f for f in os.listdir(input_dir) if f.endswith('.png') or f.endswith('.jpg')]

    for image_file in image_files:
        # Read the image
        image_path = os.path.join(input_dir, image_file)
        image = cv2.imread(image_path)

        # Crop the image using the specified coordinates
        x, y, width, height = crop_coordinates
        cropped_image = image[y:y + height, x:x + width]

        # Save the cropped image to the output directory
        output_path = os.path.join(output_dir, image_file)
        cv2.imwrite(output_path, cropped_image)

        print(f"Cropped and saved: {output_path}")

# Specify the input and output directories
input_directory = '/content/drive/MyDrive/speech emotion/Melspectro/val/angry'  # Update with your input directory path
output_directory = '/content/drive/MyDrive/speech emotion/onlyMelspectro/val/angry'  # Update with your output directory path

# Define the crop coordinates (x, y, width, height)
crop_coords = (50, 50, 200, 100)  # Update these values based on your needs

# Run the cropping function
crop_and_save_melspectrograms(input_directory, output_directory, crop_coords)
```

```python
import os
import cv2
import numpy as np

def remove_white_parts(image):
    """
    Remove white parts from the mel spectrogram image.

    Parameters:
    - image: The input image from which white parts will be removed.

    Returns:
    - The processed image with white parts removed.
    """
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply a binary threshold to create a mask for white areas
    _, mask = cv2.threshold(gray_image, 240, 255, cv2.THRESH_BINARY_INV)

    # Use the mask to keep only the non-white parts of the image
    result = cv2.bitwise_and(image, image, mask=mask)

    return result

def process_and_save_melspectrograms(input_dir, output_dir):
    """
    Process mel spectrograms by removing white parts and save them to a new directory.

    Parameters:
    - input_dir: Directory containing the original images.
    - output_dir: Directory where processed images will be saved.
    """
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Get a list of all image files in the input directory
    image_files = [f for f in os.listdir(input_dir) if f.endswith('.png') or f.endswith('.jpg')]

    for image_file in image_files:
        # Read the image
        image_path = os.path.join(input_dir, image_file)
        image = cv2.imread(image_path)

        # Remove white parts from the image
        processed_image = remove_white_parts(image)

        # Save the processed image to the output directory
        output_path = os.path.join(output_dir, image_file)
        cv2.imwrite(output_path, processed_image)

        print(f"Processed and saved: {output_path}")

# Specify the input and output directories
input_directory = '/content/drive/MyDrive/speech emotion/Melspectro/val/angry'  # Update with your input directory path
output_directory = '/content/drive/MyDrive/speech emotion/whiteonlyMelspectro/val/angry'  # Update with your output directory path

# Run the processing function
process_and_save_melspectrograms(input_directory, output_directory)
```
